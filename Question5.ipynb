## Question 5: Plot Tesla Stock **Graph**

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tesla = yf.Ticker('TSLA')
```

```python
tesla_info = tesla.history(period='max')
```

```python
tesla_revenue = tesla.financials.transpose()
tesla_revenue.reset_index(inplace=True)
tesla_revenue['Date'] = pd.to_datetime(tesla_revenue['index'])
tesla_revenue['Total Revenue'] = tesla_revenue['Total Revenue'].astype(float)
```

```python
tesla_info.reset_index(inplace=True)
```

```python
def make_graph(stock_data, revenue_data=None, plot_open=True, stock_name='Stock'):
    plt.figure(figsize=(10, 6))
    if plot_open:
        plt.plot(stock_data['Date'], stock_data['Open'], label=f'{stock_name} Opening Price', color='blue')
        plt.ylabel('Opening Price ($)')
    if revenue_data is not None:
        plt.plot(revenue_data['Date'], revenue_data['Total Revenue'], label=f'{stock_name} Revenue', color='red')
        plt.ylabel('Revenue ($)')

    plt.title(f'{stock_name} Financial Overview')
    plt.xlabel('Date')
    plt.legend()
    plt.show()
```

```python
make_graph(tesla_info, plot_open=True, stock_name='Tesla')
```

```python
make_graph(stock_data=None, revenue_data=tesla_revenue, plot_open=False, stock_name='Tesla')
```
