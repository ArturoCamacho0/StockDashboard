## Question 5: Plot Tesla Stock **Graph**

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tesla = yf.Ticker('TSLA')
```

```python
tesla_info = tesla.history(period='max')
```

```python
tesla_info.reset_index(inplace=True)
```

```python
def make_graph(stock_data, stock_name, title):
  stock_data.reset_index(inplace=True)
  plt.plot(stock_data['Date'], stock_data['Open'], label=stock_name)
  plt.title(title)
  plt.xlabel('Date')
  plt.ylabel('Opening Price')
  plt.legend()
  plt.show()
```

```python
make_graph(tesla_info, 'Tesla', 'Tesla Stock Price')
```
