## Question 6: Plot GameStop Stock Graph

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gamestop = yf.Ticker('GME')
```

```python
gme_info = gamestop.history(period='max')
```

```python
gme_revenue = gamestop.financials.transpose()
gme_revenue.reset_index(inplace=True)
gme_revenue['Date'] = pd.to_datetime(gme_revenue['index'])
gme_revenue['Total Revenue'] = gme_revenue['Total Revenue'].astype(float)
```

```python
gme_info.reset_index(inplace=True)
```

```python
def make_graph(stock_data, revenue_data=None, plot_open=True, stock_name='Stock'):
    plt.figure(figsize=(10, 6))
    if plot_open:
        plt.plot(stock_data['Date'], stock_data['Open'], label=f'{stock_name} Opening Price', color='blue')
        plt.ylabel('Opening Price ($)')
    if revenue_data is not None:
        plt.plot(revenue_data['Date'], revenue_data['Total Revenue'], label=f'{stock_name} Revenue', color='red')
        plt.ylabel('Revenue ($)')

    plt.title(f'{stock_name} Financial Overview')
    plt.xlabel('Date')
    plt.legend()
    plt.show()
```

```python
make_graph(gme_info, plot_open=True, stock_name='GameStop')
```

```python
make_graph(stock_data=None, revenue_data=gme_revenue, plot_open=False, stock_name='GameStop')
```
